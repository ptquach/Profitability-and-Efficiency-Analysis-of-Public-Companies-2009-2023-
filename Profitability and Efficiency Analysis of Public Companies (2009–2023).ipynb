```python
import pandas as pd
```

```python
df = pd.read_csv("Financial Statements.csv")
df.head()
```

```python
df.dtypes
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()
```

```python
 averageROE = df.groupby('Company')['ROE'].mean().sort_values(ascending=False)
```

```python
print("The highest average ROE companies are:\n")
print(averageROE)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='ROE', hue='Company', marker='o')

plt.title('ROE over the years')
plt.ylabel('Return of Equity (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Net Profit Margin', hue='Company', marker='o')

plt.title('Net Profit Margin Trend')
plt.ylabel('Net Profit Margin (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
 averageROA = df.groupby('Company')['ROA'].mean().sort_values(ascending=False)
```

```python
print("The highest average ROA companies are:\n")
print(averageROA)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='ROA', hue='Company', marker='o')

plt.title('ROA over the years')
plt.ylabel('Return of Assets (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
averageROI = df.groupby('Company')['ROI'].mean().sort_values(ascending=False)
```

```python
print("The highest average ROI companies are:\n")
print(averageROI)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='ROI', hue='Company', marker='o')

plt.title('ROI over the years')
plt.ylabel('Return of Investment (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Profitability Score'] = df[['ROE', 'ROA', 'ROI', 'Net Profit Margin']].mean(axis=1)
print(df['Profitability Score'])
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(df['Debt/Equity Ratio'], df['Profitability Score'], alpha=0.7)
plt.xlabel('Debt/Equity Ratio')
plt.ylabel('Profitabilitys Score')
plt.title('Leverage vs Profitability')
plt.grid(True)
plt.show()
```

```python
df[['Inflation Rate(in US)', 'ROE', 'ROA', 'ROI', 'Net Profit Margin']].corr()
```

```python
sns.heatmap(
    df[['Inflation Rate(in US)', 'ROE', 'ROA', 'ROI', 'Net Profit Margin']].corr(),
    annot=True, cmap='coolwarm'
)
plt.title("Correlation between Inflation and Profitability Metrics")
plt.show()
```

```python
df[['Inflation Rate(in US)', 'Profitability Score']].corr()
```

```python
df_avg = df.groupby('Year')[['Inflation Rate(in US)', 'Profitability Score']].mean().reset_index()

import matplotlib.pyplot as plt

plt.plot(df_avg['Year'], df_avg['Inflation Rate(in US)'], label='Inflation', color='gray', linestyle='--')
plt.plot(df_avg['Year'], df_avg['Profitability Score'], label='Avg Profitability Score', marker='o')

plt.title('Inflation vs Profitability Score (Avg Across Companies)')
plt.xlabel('Year')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['Inflation Rate(in US)'], df['Profitability Score'], alpha=0.6)
plt.title('Inflation vs Profitability Score (All Companies, All Years)')
plt.xlabel('Inflation Rate (%)')
plt.ylabel('Profitability Score')
plt.grid(True)
plt.show()
```
